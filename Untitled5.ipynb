                                     BURSA HAVA DURUMUNUN GEÇMİŞİ VE GELECEĞİ


Bursa, Türkiye'nin Marmara Bölgesi'nde yer alan tarihi ve kültürel öneme sahip bir şehirdir. Bursa'nın iklimi genel olarak karasal iklim ile Akdeniz iklimi arasında bir geçiş iklimidir. Bu da mevsimlere göre değişen sıcaklık ve yağış özellikleri anlamına gelir. 

Bu konu hakkında mevsimlere göre küçük bir tanımlama bursa hakkında bize yardımcı olacak:

Kış (Aralık - Şubat): Kış aylarında Bursa soğuk ve yağışlı bir dönem geçirir. Kar yağışı olabilir, özellikle dağlık bölgelerde (Uludağ) daha fazla kar görülebilir. Ortalama sıcaklıklar genellikle 0°C ile 10°C arasında değişir.

İlkbahar (Mart - Mayıs): İlkbaharda hava ısınmaya başlar ve bitki örtüsü canlanır. Bu dönemde hava sıcaklıkları artar, ancak yağışlar da devam edebilir. İlkbahar aylarında ortalama sıcaklıklar 10°C ile 20°C arasında değişir.

Yaz (Haziran - Ağustos): Yaz ayları Bursa'da sıcak ve kurudur. Sıcaklık genellikle 25°C ile 35°C arasında değişir. Bu dönemde yağış miktarı azalır ve nem düşer. Özellikle Temmuz ve Ağustos aylarında sıcaklık daha yüksektir.

Sonbahar (Eylül - Kasım): Sonbahar aylarında hava tekrar ılımanlaşmaya başlar. Sıcaklıklar düşmeye başlar ancak hala keyifli bir dönemdir. Yağışlar artmaya başlar. Eylül ve Ekim aylarında sıcaklıklar 15°C ile 25°C arasında değişir.

Kullanacağımız Parametreleri Kısaca Tanıyalım:

Sıcaklık: Havanın sıcaklığını belirtir. Genellikle Celsius veya Fahrenheit cinsinden ifade edilir.

Nem: Havanın içerdiği su buharı miktarını ifade eder. Genellikle yüzde (%) cinsinden verilir.

Yağış: Belirli bir zaman diliminde düşen yağmur, kar veya diğer yağış türlerinin miktarını ifade eder. Milimetre veya inç cinsinden verilebilir.

ŞİMDİ DAHA ÇOK VERİYE GİRİŞ YAPALIM

Bursanın geçtiğimiz 5 yılda ki ortalama hava durumu verilerine mevsimsel olarak bakalım:

```python
import matplotlib.pyplot as plt

# Verileri hazırla
mevsimler = ['Bahar', 'Yaz', 'Sonbahar', 'Kış']
ortalama_sicaklik = [18, 25.5, 15.5, 5]
ortalama_yagis = [216, 80, 226, 396]
nem_oranlari = [25, 37, 18, 10]

# Renk paleti
renkler = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']

# Dairesel grafik oluştur
plt.figure(figsize=(10, 8))

# Alt grafikler oluşturarak 3 değeri birden gösterelim
plt.subplot(1, 3, 1)
plt.pie(ortalama_sicaklik, labels=mevsimler, autopct='%1.1f%%', colors=renkler, startangle=140)
plt.title('Ortalama Sıcaklık')

plt.subplot(1, 3, 2)
plt.pie(ortalama_yagis, labels=mevsimler, autopct='%1.1f%%', colors=renkler, startangle=140)
plt.title('Ortalama Yağış')

plt.subplot(1, 3, 3)
plt.pie(nem_oranlari, labels=mevsimler, autopct='%1.1f%%', colors=renkler, startangle=140)
plt.title('Nem Oranları')

plt.tight_layout()  # Grafiklerin örtüşmesini engellemek için
plt.suptitle('Bursa Hava Durumu Değerleri Mevsimlere Göre', fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Verileri hazırla
mevsimler = ['Bahar', 'Yaz', 'Sonbahar', 'Kış']
ortalama_sicaklik = [18, 25.5, 15.5, 5]
ortalama_yagis = [216, 80, 226, 396]
nem_oranlari = [25, 37, 18, 10]

# Renkler
renk_ortalama_sicaklik = 'red'
renk_ortalama_yagis = 'blue'
renk_nem_oranlari = 'green'

# Grafik oluştur
plt.figure(figsize=(10, 6))

# Ortalama Sıcaklık grafiği
plt.plot(mevsimler, ortalama_sicaklik, marker='o', color=renk_ortalama_sicaklik, label='Ortalama Sıcaklık')

# Ortalama Yağış grafiği
plt.plot(mevsimler, ortalama_yagis, marker='o', color=renk_ortalama_yagis, label='Ortalama Yağış')

# Nem Oranları grafiği
plt.plot(mevsimler, nem_oranlari, marker='o', color=renk_nem_oranlari, label='Nem Oranları')

# Eksen ve başlık
plt.xlabel('Mevsim')
plt.ylabel('Değer')
plt.title('Bursa Hava Durumu Değerleri Mevsimlere Göre')
plt.legend()

# Otomatik şekilde düzenle
plt.tight_layout()

# Grafik göster
plt.show()
```

ŞİMDİ BİRAZDA GELECEK HAVA DURUMLARI HAKKINDA TAHMİNSEL DURUMLAR HAKKINDA KONUŞALIM

Bursa'nın iklimi, Türkiye'nin Marmara Bölgesi'nde yer aldığı için kara iklimi ve Akdeniz iklimi arasında bir geçiş iklimi özelliği gösterir. Ancak iklim değişikliği gibi küresel etkiler, Bursa'nın gelecekteki iklim durumunu etkileyebilir. Gelecekte Bursa'nın iklimi hakkında öngörülen bazı potansiyel değişiklikler şunlar olabilir:

Artan Sıcaklıklar: İklim değişikliği nedeniyle sıcaklıkların artması bekleniyor. Bu, yaz aylarında daha yüksek sıcaklıkların yaşanması anlamına gelebilir. Bu durum, sıcaklıkla ilgili sorunları artırabilir ve su kaynaklarını etkileyebilir.

Yağış Değişiklikleri: Bursa'nın yağış desenleri değişebilir. Daha az yağış veya yoğun yağış olayları yaşanabilir. Bu, su kaynaklarının yönetimini etkileyebilir, kuraklık ve sel riskini artırabilir.

Su Kaynakları Sorunları: Daha sıcak ve kuru koşullar, su kaynaklarının azalmasına ve su sıkıntısına yol açabilir. Bu, tarım, içme suyu temini ve sanayi için ciddi sorunlar yaratabilir.

Bitki Örtüsü ve Ekosistem Değişiklikleri: İklim değişikliği, bitki örtüsü ve ekosistemleri etkileyebilir. Bazı bitki ve hayvan türleri, değişen koşullara uyum sağlayamayabilir.

Sıcak Ada Etkisi: Büyük şehirlerde sıcak ada etkisi daha fazla hissedilir hale gelebilir. Bu, kent içi sıcaklıkların artmasına ve yaşam koşullarının zorlaşmasına yol açabilir.

Tarım ve Gıda Güvenliği: İklim değişikliği, tarım üretimini etkileyebilir. Değişen sıcaklık ve yağış koşulları, ürün verimini ve kalitesini etkileyebilir.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Geçmiş verileri oluştur
years = np.array([2000, 2001, 2002, 2003, 2004, 2005])
data = np.array([2, 3, 4, 6, 7, 8])  # Sadece örnek veri, gerçek verilerinizle değiştirin

# Verileri yeniden şekillendir
years = years.reshape(-1, 1)
data = data.reshape(-1, 1)

# Doğrusal regresyon modeli oluştur
model = LinearRegression()
model.fit(years, data)

# Tahminler yapmak için gelecek yılları oluştur
future_years = np.array([2025, 2026, 2027, 2028, 2029]).reshape(-1, 1)

# Gelecek yıllar için tahmin yap
predicted_data = model.predict(future_years)

# Tahmin sonuçlarını çizdir
plt.scatter(years, data, color='blue', label='Geçmiş Veriler')
plt.plot(future_years, predicted_data, color='red', label='Tahmin')
plt.xlabel('Yıl')
plt.ylabel('Veri')
plt.title('Geçmiş Verilere Dayalı Tahmin')
plt.legend()
plt.show()
```
